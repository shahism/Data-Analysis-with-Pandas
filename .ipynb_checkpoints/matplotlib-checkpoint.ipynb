```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3,4,5],[0,1,1.5,2.0,2.5])
plt.title("Shah's Data")
plt.xlabel("X data")
plt.ylabel("Y data")
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot")
plt.plot([1,2,3,4,5],[0,1,1.5,2.0,2.5])
plt.title("Shah's Data")
plt.xlabel("X data")
plt.ylabel("Y data")
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
x = [2,5,8]
y = [22,55,77]

x1 = [1,7,9]
y1 = [2,9,15]

plt.plot(x,y ,label="first" , linewidth= 4)
plt.plot(x1,y1, label = "Second")
plt.title("Shah's Data")
plt.legend() #legend is called for displaying the labels
plt.show() 
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
x = [2,5,8]
y = [22,55,77]

x1 = [1,7,9]
y1 = [2,9,15]

plt.scatter(x,y ,label="first" , linewidth= 4)
plt.scatter(x1,y1, label = "Second")
plt.title("Shah's Data")
plt.legend() #legend is called for displaying the labels
plt.show() 
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
x = [100,50,8]
y = [22,55,77]

x1 = [1,7,9]
y1 = [2,9,15]

plt.hist(x)
plt.plot(x1,y1, label = "Second")
plt.title("Shah's Data")
plt.legend() #legend is called for displaying the labels
plt.show() 


# histogram can be plotted using this matplotlib 
# baaki documentation se hoga jo krna hai . 
```
